```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime as dt
import zipfile
```

```python
xivisit = pd.read_excel('./xivisit.xlsx')
xivisit.info()
```

```python
industry_group = xivisit.groupby(['year', 'industry']).size().reset_index(name='count')
```

```python

list_group = xivisit.groupby(['year', 'listed']).size().reset_index(name='count')
list_group.sample(10)
list_group = list_group.pivot(index='year', columns='listed', values='count')
```

```python
# Step 3: Plot the data
list_group.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Listed Company per Year')
plt.legend(title='listed')
plt.show();
```

```python
pivot_data = industry_group.pivot(index='year', columns='industry', values='count')

# Step 3: Plot the data
pivot_data.plot(kind='line')
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Count by Industry per Year')
plt.legend(title='Industry')
plt.show();
```
